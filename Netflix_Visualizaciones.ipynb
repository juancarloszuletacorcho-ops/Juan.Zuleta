```python
import plotly.express as px

fig = px.histogram(df, x="type", color="type",
                   title="Distribución del contenido por formato (Movie vs TV Show)")
fig.show()
```

```python
df["date_added"] = pd.to_datetime(df["date_added"], errors="coerce")
df["year_added"] = df["date_added"].dt.year

trend = df.groupby(["year_added","type"]).size().reset_index(name="count")

fig = px.line(trend, x="year_added", y="count", color="type",
              title="Evolución del contenido agregado por formato")
fig.show()
```

```python
genres = df["listed_in"].dropna().str.split(",").explode().str.strip()
top_genres = genres.value_counts().head(15).reset_index()
top_genres.columns = ["Genero", "Cantidad"]

fig = px.bar(top_genres, x="Genero", y="Cantidad",
             title="Top 15 géneros audiovisuales en Netflix")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
countries = df["country"].dropna().str.split(",").explode().str.strip()
top_countries = countries.value_counts().head(15).reset_index()
top_countries.columns = ["Pais", "Cantidad"]

fig = px.bar(top_countries, x="Pais", y="Cantidad",
             title="Top 15 países con mayor número de títulos en Netflix")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
format_counts = df["type"].value_counts()

fig = px.pie(values=format_counts.values,
             names=format_counts.index,
             title="Proporción del catálogo por formato",
             hole=0.4)  # donut, más elegante
fig.show()
```

```python
directors = df["director"].dropna().str.split(",").explode().str.strip()
top_directors = directors.value_counts().head(15).reset_index()
top_directors.columns = ["Director","Cantidad"]

import plotly.express as px

fig = px.bar(top_directors, x="Director", y="Cantidad",
             title="Top 15 directores más frecuentes en Netflix")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
df["year_added"] = pd.to_datetime(df["date_added"], errors="coerce").dt.year
recent = df[df["year_added"] >= 2018]

directors_recent = recent["director"].dropna().str.split(",").explode().str.strip()
top_recent = directors_recent.value_counts().head(15).reset_index()
top_recent.columns = ["Director","Cantidad"]

fig = px.bar(top_recent, x="Director", y="Cantidad",
             title="Top directores en tendencia reciente (desde 2018)")
fig.update_layout(xaxis_tickangle=-45)
fig.show()
```

```python
directores_clave = top_recent["Director"].head(5).tolist()

df_dir = df.dropna(subset=["director"]).assign(
    director=df["director"].str.split(",")
).explode("director")
df_dir["director"] = df_dir["director"].str.strip()

df_dir = df_dir[df_dir["director"].isin(directores_clave)]

trend_dir = df_dir.groupby(["year_added","director"]).size().reset_index(name="count")

fig = px.line(trend_dir, x="year_added", y="count", color="director",
              title="Evolución temporal de directores clave")
fig.show()
```

```python
df_cast = df.dropna(subset=["cast","country"]).copy()

df_cast["cast"] = df_cast["cast"].str.split(",")
df_cast["country"] = df_cast["country"].str.split(",")

df_cast = df_cast.explode("cast").explode("country")

df_cast["cast"] = df_cast["cast"].str.strip()
df_cast["country"] = df_cast["country"].str.strip()
```

```python
actor_country = df_cast.groupby("country").size().reset_index(name="count")
top_actor_countries = actor_country.sort_values("count", ascending=False).head(15)
```
