```python
# ==========================================================
# PROYECTO: PREDICCIÓN DE APROBACIÓN DE PRÉSTAMOS
# VERSIÓN CORREGIDA - 100% FUNCIONAL EN GOOGLE COLAB
# ==========================================================

# ===============================
# 1. IMPORTAR LIBRERÍAS
# ===============================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import (
    accuracy_score,
    classification_report,
    confusion_matrix,
    roc_auc_score,
    RocCurveDisplay
)

# ===============================
# 2. GENERAR BASE DE DATOS REALISTA
# (Evita errores del CSV y clases únicas)
# ===============================
np.random.seed(42)

n = 120

edad = np.random.randint(21, 61, n)
ingresos = np.random.randint(1500, 9000, n)
score = np.random.randint(300, 851, n)
prestamos_previos = np.random.randint(0, 6, n)

# Fórmula lógica de aprobación
probabilidad = (
    (ingresos / 9000) * 0.35 +
    (score / 850) * 0.45 +
    ((60 - prestamos_previos) / 60) * 0.20
)

categoria = []

for p in probabilidad:
    if p > 0.62:
        categoria.append(1)
    else:
        categoria.append(0)

# Crear DataFrame
df = pd.DataFrame({
    "Edad": edad,
    "Ingresos_Mensuales": ingresos,
    "Puntuacion_Crediticia": score,
    "Prestamos_Previos": prestamos_previos,
    "Categoria_Prestamo": categoria
})

# Guardar CSV
df.to_csv("datos_prestamos.csv", index=False)

print("Archivo creado correctamente")
print(df.head())

# ===============================
# 3. EXPLORACIÓN DE DATOS
# ===============================
print("\nINFORMACIÓN GENERAL")
print(df.info())

print("\nESTADÍSTICAS")
print(df.describe())

print("\nDISTRIBUCIÓN CLASES")
print(df["Categoria_Prestamo"].value_counts())

# ===============================
# 4. VARIABLES X / y
# ===============================
X = df.drop("Categoria_Prestamo", axis=1)
y = df["Categoria_Prestamo"]

# ===============================
# 5. DIVIDIR TRAIN / TEST
# ===============================
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.25,
    random_state=42,
    stratify=y
)

# ===============================
# 6. NORMALIZACIÓN
# ===============================
scaler = StandardScaler()

X_train_sc = scaler.fit_transform(X_train)
X_test_sc = scaler.transform(X_test)

# ===============================
# 7. MODELO REGRESIÓN LOGÍSTICA
# ===============================
modelo = LogisticRegression(max_iter=1000)

modelo.fit(X_train_sc, y_train)

# ===============================
# 8. PREDICCIONES
# ===============================
pred = modelo.predict(X_test_sc)
proba = modelo.predict_proba(X_test_sc)[:,1]

# ===============================
# 9. EVALUACIÓN
# ===============================
print("\nACCURACY:")
print(accuracy_score(y_test, pred))

print("\nREPORTE CLASIFICACIÓN:")
print(classification_report(y_test, pred))

# ===============================
# 10. MATRIZ CONFUSIÓN
# ===============================
cm = confusion_matrix(y_test, pred)

plt.figure(figsize=(6,4))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Matriz de Confusión")
plt.xlabel("Predicho")
plt.ylabel("Real")
plt.show()

# ===============================
# 11. CURVA ROC
# ===============================
print("ROC AUC:", roc_auc_score(y_test, proba))

RocCurveDisplay.from_predictions(y_test, proba)
plt.show()

# ===============================
# 12. NUEVOS CLIENTES
# ===============================
nuevo_cliente = pd.DataFrame({
    "Edad": [35, 52],
    "Ingresos_Mensuales": [4200, 7800],
    "Puntuacion_Crediticia": [640, 790],
    "Prestamos_Previos": [2, 0]
})

nuevo_sc = scaler.transform(nuevo_cliente)

resultado = modelo.predict(nuevo_sc)
prob = modelo.predict_proba(nuevo_sc)[:,1]

nuevo_cliente["Prediccion"] = resultado
nuevo_cliente["Probabilidad_Aprobacion"] = prob

print("\nNUEVOS CLIENTES:")
print(nuevo_cliente)
```
